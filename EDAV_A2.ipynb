```python
# #Q1. Plot a line chart showing the average user rating per year
# for Fiction and Non-Fiction books separately. Identify whether
# one genre consistently outrates the other.

# importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# loading dataset
df = pd.read_csv("bestsellers with categories.csv")

# grouping data to get average rating per year and genre
avg_rating = df.groupby(['Year', 'Genre'])['User Rating'].mean().reset_index()

# separating fiction and non-fiction
fiction = avg_rating[avg_rating['Genre'] == 'Fiction']
non_fiction = avg_rating[avg_rating['Genre'] == 'Non Fiction']

# plotting
plt.figure(figsize=(10,6))

plt.plot(fiction['Year'], fiction['User Rating'], marker='o', color='blue', label='Fiction')
plt.plot(non_fiction['Year'], non_fiction['User Rating'], marker='o', color='green', label='Non-Fiction')

# labels and title
plt.title("Average User Rating per Year for Fiction and Non-Fiction Books")
plt.xlabel("Year")
plt.ylabel("Average User Rating")

# legend
plt.legend(title="Genre")

# grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# show plot
plt.show()
```

**Analysis**

The line chart shows the average user ratings for Fiction and Non-Fiction books from 2009 to 2019.

**Key Observations:**

*   **High Ratings:** Both genres consistently have high user ratings, generally between 4.5 and 4.9. This indicates strong reader satisfaction.

*   **Fiction Generally Higher:** Upon closer inspection, Fiction books tend to show slightly higher average ratings compared to Non-Fiction in most of the years (2009, 2010, 2011, 2017, 2018, and 2019). Non-Fiction had slightly higher ratings in 2012 and 2016. The differences are generally small (less than 0.2 points).

*   **Stable Trends:** Ratings for both genres are quite stable over the decade, with minor year-to-year fluctuations. Fiction shows a bit more variability, with a notable increase in ratings from 2016 to 2019.

**Conclusion:**

While both Fiction and Non-Fiction genres are highly rated, Fiction generally holds a slight edge in average user ratings over the 2009-2019 period, appearing higher in more years. Both categories feature books that resonate well with their audiences.

```python
# Create a scatter plot of Price vs. Reviews for all books.
# Color points by Genre. Use log scale for Reviews. Identify high-
# value, high-engagement outliers.
# creating scatter plot
plt.figure(figsize=(10,6))

# plotting fiction
fiction = df[df['Genre'] == 'Fiction']
plt.scatter(fiction['Price'], fiction['Reviews'],
            color='blue', label='Fiction', alpha=0.7)

# plotting non-fiction
non_fiction = df[df['Genre'] == 'Non Fiction']
plt.scatter(non_fiction['Price'], non_fiction['Reviews'],
            color='orange', label='Non-Fiction', alpha=0.7)

# applying log scale to Reviews
plt.yscale('log')

# labels and title
plt.xlabel("Price of Book")
plt.ylabel("Number of Reviews (log scale)")
plt.title("Scatter Plot of Price vs Reviews for Books")

# legend
plt.legend(title="Genre")

# grid
plt.grid(True, linestyle='--', alpha=0.5)

# show plot
plt.show()
```

**Analysis**

This scatter plot shows book price against review count (log scale), colored by genre.

**Key Observations:**

*   **Price Distribution:** Most books, for both genres, are concentrated in lower to mid-price ranges ($0-$30).

*   **Review Engagement:** High review counts are common across various price points and in both genres, indicating strong reader engagement.

*   **Genre Comparison:** Non-Fiction shows a slightly wider distribution into higher price categories. Both genres have many highly-reviewed books.

*   **Outliers:** Books with high prices and numerous reviews represent 'high-value, high-engagement' outliers, visible in both genres, with some notable candidates in Non-Fiction at higher prices.

**Conclusion:**

High reader engagement is present in both Fiction and Non-Fiction across various price points. While most bestsellers are affordable, a market exists for premium content that still garners significant reader interaction.

```python
# Visualize the distribution of book prices using a histogram.
# Overlay KDE curves for Fiction and Non-Fiction. Identify the
# most common price point for each genre.
import seaborn as sns
plt.figure(figsize=(10,6))

# histogram with KDE for both genres
sns.histplot(data=df, x='Price', hue='Genre', kde=True, bins=20, alpha=0.5)

# labels and title
plt.xlabel("Book Price")
plt.ylabel("Frequency")
plt.title("Distribution of Book Prices with KDE for Fiction and Non-Fiction")



# grid for clarity
plt.grid(True, linestyle='--', alpha=0.5)

# show plot
plt.show()
```

**Analysis**

The histogram with overlaid Kernel Density Estimate (KDE) curves illustrates the distribution of book prices for Fiction and Non-Fiction genres. The x-axis represents the book price, and the y-axis shows the frequency (or density for KDE) of books at those price points.

**Key Observations:**

*   **Right-Skewed Distribution:** Both Fiction and Non-Fiction price distributions are heavily right-skewed, meaning a large majority of books are priced at the lower end, with fewer books at higher price points.

*   **Most Common Price Point (Mode):**
    *   **Fiction:** The KDE curve for Fiction peaks sharply around a price of $8-$10. This indicates that the most common price point for Fiction bestsellers is in this range.
    *   **Non-Fiction:** The KDE curve for Non-Fiction also shows a prominent peak, slightly broader than Fiction, primarily in the $8-$12 range. It also has a smaller, secondary peak around $20-$25, suggesting a segment of Non-Fiction books at a moderately higher price.

*   **Genre Comparison:**
    *   While both genres share a similar primary most common price point, Non-Fiction shows a slightly wider spread and often has a greater number of books in the mid-to-higher price ranges (e.g., above $15-$20) compared to Fiction, which drops off more rapidly after its peak.
    *   Fiction's distribution is more tightly clustered around its dominant mode.

**Conclusion:**

The majority of best-selling books in both Fiction and Non-Fiction are affordably priced, with the most common prices falling between $8 and $12. Non-Fiction, however, tends to have a more diverse price structure, with a notable portion of its books selling at moderately higher prices than Fiction, suggesting a broader market segment for pricier Non-Fiction content.

```python
# Create a density plot of User Rating grouped by year
# (grouped into periods: 2009–2013, 2014–2019). Identify whether
# average quality shifted across periods.

# creating time periods
def assign_period(year):
    if 2009 <= year <= 2013:
        return "2009-2013"
    else:
        return "2014-2019"

df['Period'] = df['Year'].apply(assign_period)

# plotting density plot
plt.figure(figsize=(10,6))

sns.kdeplot(data=df, x='User Rating', hue='Period', fill=True, alpha=0.5)

# labels and title
plt.xlabel("User Rating")
plt.ylabel("Density")
plt.title("Density Plot of User Ratings by Time Period")

# legend
plt.legend(title="Time Period")

# grid
plt.grid(True, linestyle='--', alpha=0.5)

# show plot
plt.show()
```

**Analysis**

The Upward Shift in User Ratings
Higher Peak (The "Mode"): There is a clear rightward shift in the data. In the earlier period (2009–2013), the most common rating was around 4.6. In the later period (2014–2019), that peak jumped to 4.8, showing that bestsellers are being rated significantly higher than they used to be.

Greater Consensus: The blue curve (2014–2019) is much taller and narrower. This indicates that ratings became more "concentrated" in the later years. There is less variation; almost every bestseller is now expected to have a near-perfect score.

Vanishing Low Ratings: The "tail" on the left is much smaller for the later period. In the early years, you could still find bestsellers with ratings around 4.0, but by 2014–2019, those have almost entirely disappeared from the Top 50 list.

Conclusion: The "average quality" has shifted upward. Whether the books actually got better or people just became more generous with 5-star reviews, a book now needs a higher rating than ever to compete for a bestseller spot.

```python
# Use a 2D density contour plot (sns.kdeplot) of Price vs.
# Reviews. Identify clusters of popular affordable vs. expensive
# underperforming books.

import seaborn as sns
# setting figure size
plt.figure(figsize=(10,6))

# 2D density contour plot
sns.kdeplot(
    data=df,
    x='Price',
    y='Reviews',
    fill=True,
    cmap='viridis',
    levels=15, # Increased levels for better detail
    thresh=0.01 # Lowered thresh to see more data
)

# Crucial: Set limits so we can actually see the clusters
plt.xlim(0, 60) # Most books are under $60
plt.ylim(0, 40000) # Most books have under 40k reviews

plt.xlabel("Price of Book")
plt.ylabel("Number of Reviews")
plt.title("2D Density Contour Plot of Price vs Reviews")
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

**Analysis**

The "Sweet Spot" (Yellow Peak): The highest density of books is clustered between $8 and $12 with 5,000 to 10,000 reviews. This is the primary cluster of popular, affordable books that dominate the bestseller list.

The Price Barrier: As the price increases beyond $20, the density drops significantly (shifting to dark purple). This shows that while expensive books can be bestsellers, they are much less common and typically have fewer reviews.

The "Expensive Underperformers": In the $40–$60 range, the clusters are very low on the Y-axis. These books represent the "underperforming" segment—they are priced high but struggle to gain the massive review volume seen in cheaper genres.

The Outlier "Island": There is a unique, isolated cluster near $40 with ~35,000 reviews. This represents a rare "high-value" bestseller—an expensive book that defied the trend and gained massive popularity.

Conclusion: The data shows a strong inverse relationship between high price and high review volume. Success on the bestseller list is most "dense" when the price is kept under $15

```python
# Visualize error bars of average price per year (mean ± std).
# Determine whether pricing became more or less variable over
# the decade.
# importing libraries


# grouping data to calculate mean and std per year
price_stats = df.groupby('Year')['Price'].agg(['mean', 'std']).reset_index()

# plotting error bars
plt.figure(figsize=(10,6))

plt.errorbar(
    price_stats['Year'],
    price_stats['mean'],
    yerr=price_stats['std'],
    fmt='o-',
    capsize=5,
    color='purple',
    ecolor='gray',
    label='Mean Price ± Std Dev'
)

# labels and title
plt.xlabel("Year")
plt.ylabel("Average Book Price")
plt.title("Average Book Price per Year with Error Bars (Mean ± Std)")

# legend
plt.legend()

# grid
plt.grid(True, linestyle='--', alpha=0.5)

# show plot
plt.show()
```

**Analysis**

The error bar plot shows that average book prices fluctuate slightly but generally **decrease over the decade**. Prices are higher in earlier years and trend lower in later years.

The error bars indicate price variability was larger in some earlier years, with wider spreads, while later years show somewhat smaller variability and more consistency.

Overall, book pricing appears to have become **slightly less variable over time**, while average prices also show a modest downward trend.


```python
# Build a 2x2 subplot grid: (a) bar chart of top 10 authors by
# total reviews, (b) line plot of average price per year, (c) boxplot
# of reviews by genre, (d) scatter of rating vs. reviews.
# importing libraries


# creating subplot grid
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# ---------------- (a) Top 10 authors by total reviews ----------------
top_authors = df.groupby('Author')['Reviews'].sum().sort_values(ascending=False).head(10)

axes[0,0].bar(top_authors.index, top_authors.values, color='skyblue')
axes[0,0].set_title("Top 10 Authors by Total Reviews")
axes[0,0].set_xlabel("Author")
axes[0,0].set_ylabel("Total Reviews")
axes[0,0].tick_params(axis='x', rotation=45)

# ---------------- (b) Line plot: avg price per year ----------------
avg_price = df.groupby('Year')['Price'].mean()

axes[0,1].plot(avg_price.index, avg_price.values, marker='o', color='green')
axes[0,1].set_title("Average Price per Year")
axes[0,1].set_xlabel("Year")
axes[0,1].set_ylabel("Average Price")
axes[0,1].grid(True, linestyle='--', alpha=0.5)

# ---------------- (c) Boxplot: reviews by genre ----------------
sns.boxplot(data=df, x='Genre', y='Reviews', ax=axes[1,0])
axes[1,0].set_title("Distribution of Reviews by Genre")
axes[1,0].set_xlabel("Genre")
axes[1,0].set_ylabel("Reviews")

# ---------------- (d) Scatter: rating vs reviews ----------------
axes[1,1].scatter(df['User Rating'], df['Reviews'], alpha=0.6, color='purple')
axes[1,1].set_title("User Rating vs Reviews")
axes[1,1].set_xlabel("User Rating")
axes[1,1].set_ylabel("Reviews")

# layout adjustment
plt.tight_layout()

# show plot
plt.show()
```

**Analysis**

The subplot dashboard highlights multiple patterns in the dataset. **Suzanne Collins** has the highest total reviews among authors, indicating strong reader engagement. The average book price shows some fluctuations but generally trends downward over the years.

The boxplot suggests **Fiction** books have greater variability in reviews and more high-review outliers compared to Non-Fiction. The scatter plot shows most books maintain high ratings regardless of review count, with only a weak relationship between ratings and reviews.

Overall, pricing appears relatively stable, while popularity varies significantly across authors and genres.


```python
# Use Seaborn stripplot and boxplot (combined) to show
# user rating distribution by genre and year bucket. Identify
# outliers and rating trends.

# creating year buckets
def year_bucket(year):
    if 2009 <= year <= 2013:
        return "2009-2013"
    else:
        return "2014-2019"

df['Year_Bucket'] = df['Year'].apply(year_bucket)

# combine genre and year bucket for grouped categories
df['Group'] = df['Genre'] + " | " + df['Year_Bucket']

# plotting
plt.figure(figsize=(12,6))

# boxplot shows distribution + outliers
sns.boxplot(
    data=df,
    x='Group',
    y='User Rating',
    palette='Set2'
)

# stripplot overlays individual observations
sns.stripplot(
    data=df,
    x='Group',
    y='User Rating',
    color='black',
    alpha=0.5,
    jitter=True
)

# labels and title
plt.title("User Rating Distribution by Genre and Year Bucket")
plt.xlabel("Genre and Year Bucket")
plt.ylabel("User Rating")

# grid
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

**Analysis**

The combined boxplot and stripplot show that user ratings are generally **high and concentrated between 4.4 and 4.9** across all genre-year groups. Fiction in **2014–2019** appears slightly shifted toward higher ratings, suggesting a small improvement in average quality in the later period.

Some lower-rated books appear as outliers, particularly in **Fiction 2009–2013**, showing greater variability. Non-Fiction ratings appear relatively consistent across both periods.

Overall, ratings remain stable across genres and years, with a slight upward trend and tighter distribution in the later period.


```python
# Create a Seaborn heatmap of average user rating by Author (top 20)
# and Year. Annotate and identify the most consistently top-rated authors.
# importing libraries

# top 20 authors based on frequency in bestseller list
top_authors = df['Author'].value_counts().head(20).index

# filtering only those authors
top_df = df[df['Author'].isin(top_authors)]

# creating pivot table
rating_matrix = top_df.pivot_table(
    values='User Rating',
    index='Author',
    columns='Year',
    aggfunc='mean'
)

# plotting heatmap
plt.figure(figsize=(14,8))

sns.heatmap(
    rating_matrix,
    annot=True,
    cmap='YlGnBu',
    fmt=".2f",
    linewidths=0.5
)

plt.title("Average User Rating by Top 20 Authors and Year")
plt.xlabel("Year")
plt.ylabel("Author")

plt.show()
```

**Analysis**

The heatmap shows that most top authors maintain **consistently high ratings**, generally between **4.5 and 4.9**, indicating strong reader satisfaction over the years. Authors like **Dr. Seuss, Eric Carle, Jeff Kinney, and Sarah Young** appear consistently top-rated with very little variation.

Some authors show occasional dips or fluctuations, but overall ratings remain stable. Darker shades across entire rows indicate authors with sustained high performance.

Overall, the visualization suggests **average quality remained consistently high**, with a few authors standing out as the most consistently top-rated across multiple years.


```python
# Use Seaborn pairplot on Price, User Rating, and
# Reviews grouped by Genre. Identify the relationships that best
# separate Fiction from Non-Fiction.


# selecting required columns
pair_data = df[['Price','User Rating','Reviews','Genre']]

# creating pairplot
sns.pairplot(
    pair_data,
    hue='Genre',
    diag_kind='kde',
    plot_kws={'alpha':0.6},
    corner=True
)

plt.suptitle(
    "Pairplot of Price, User Rating and Reviews by Genre",
    y=1.02
)

plt.show()
```

**Analysis**

The pairplot shows that both Fiction and Non-Fiction books are generally low-priced and highly rated, with most ratings concentrated around **4.4–4.8**.

There is **no strong relationship between price and user rating**, as highly rated books appear across all price ranges. The **Price vs Reviews** relationship shows that many popular books are affordable and receive high review counts, while a few expensive books have lower engagement.

The clearest separation between genres appears in **review patterns**, where Non-Fiction is slightly more concentrated and Fiction shows more spread. Overall, **Reviews vs Price** provides the best distinction between the two genres, while ratings remain consistently high for both.
